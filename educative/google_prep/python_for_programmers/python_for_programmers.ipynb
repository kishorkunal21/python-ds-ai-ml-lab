```python
def factor(num : int):
    factor_list = []
    for i in range(1, num+1):
        if(num % i ==0):
            factor_list.append(i)
    print(f' factor of {num} : {factor_list}')
    return factor_list

factor(10)
```

```python
def prime_check(nums : list[int]):
    for x in nums:
        if set(factor(x)) == set([1,x]):
            print(f' is PRIME')
        else:     
            print(f' is NOT PRIME')

prime_check([1,2,3,4,5,6,7])
```

```python
def get_HCF(num1,num2):
    x = 1
    HCF = []
    while(x <= min(num1,num2)):
        if num1 % x == 0 and num2 % x == 0:
            HCF.append(x)
        x = x+1
    
    print(HCF, f'HCF of {num1},{num2} is {HCF[-1]}')
    return HCF[-1]

get_HCF(4,8)
```

#### Nested loops



```python
def ordered_pair(num1:int):
    for i in range(1, num1):
        for j in range(1, num1):
            print(f'({i}, {j})',end=" ")
        print('\n')

ordered_pair(5)    
```

```python
#Square of asterisks

def square_of_astericks(num:int):
    for i in range(1,num+1):
        print(f"{'*'*num}")

square_of_astericks(5)
```

```python
def right_triangle(num:int):
    for i in range(1,num+1):
        print('*'*i)

right_triangle(5)
```

```python
def hollow_square(num:int):
    for i in range(1,num+1):
        if(i == 1 or i == num):
            print('*' * num)    
        else:
            print(f'*{' ' * (num-2)}*')

hollow_square(5)
```

#### Function Scope
- Learn about variable scope, global variables, and what happens when we pass mutable and immutable variables as function parameters.

```python
name = "kk"

def fun():
    name = "kk1"

fun()

print(name)# The value of 'name' remains unchanged.
```

```python
## global variables

name = "kk"


def fun1():
    print('1st') 
    name = "kk1"

def fun():
    global name
    name = "kk1"

name = "kk2"
fun1()
print(name)
fun()

print(name)# The value of 'name' CHANGED

```

#### functions as arguments

```python
def add(a,b):
    return a + b

def sub(a,b):
    return a - b

def mul(a,b):
    return a * b

def div(a,b):
    return a / b

def calc(operation, a,b):
    return operation(a,b)

print(calc(add,1,2))
```

```python
add = lambda a,b: a + b

sub = lambda a,b : a - b

mul = lambda a,b : a * b

div = lambda a,b : a / b

def calc(operation, a,b):
    return operation(a,b)

print(calc(add,1,2))
```

```python
def calc(operation, a,b):
    return operation(a,b)

print(calc(lambda n1,n2 : n1 + n2,1,2))
```

#### The map object
- The built-in map() function in Python is used to apply a given function to all items in an input list (or any iterable), creating a map object. This map object is an iterator that generates results on the fly and can be converted into a list or another iterable type. The map() function returns a map object, which is an iterator. This map object can be converted to a list using list() or to another iterable type, such as a tuple, using tuple().

```python
num_list = [1,2,3,4,5]

num_list = map(lambda a : a *10,num_list)
num_list = list(num_list)
num_list
```

#### The reduce object
- The reduce() function in Python returns the final result of applying a function cumulatively to the items of an iterable, not a "reduce object" like map() or filter() return their own respective objects.

```python
from functools import reduce

#reduce(function, iterable[, initializer])

reduce(lambda x,y:x+y,num_list)
```

#### The filter object
- Another similar example is the filter() a function that returns a new object that filters the items in an iterable based on a function that returns either True or False without changing the original list. It requires a function and a list. filter() filters elements from a list of the elements satisfy the condition that is specified in the argument function. Just like map(), filter() returns a new object without changing the original list.

Syntax for filter()
filter(function, iteratable)


```python
list(filter(lambda x :x>30,num_list))
```

```python
#square each no using map
print(list(map(lambda x:x**2, num_list)))
print(num_list)
```

```python
list(filter(lambda a : a<30,num_list))
```
